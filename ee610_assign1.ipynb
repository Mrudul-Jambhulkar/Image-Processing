```python
nimg = mpimg.imread('night.jpg')
nimgc = np.copy(nimg)

# Output Images
plt.imshow(nimg)
```

```python
#https://www.geeksforgeeks.org/python-grayscaling-of-images-using-opencv/
img = cv2.imread('night.jpg')

# Obtain the dimensions of the image array
# using the shape method
(row, col) = img.shape[0:2]

# Take the average of pixel values of the BGR Channels
# to convert the colored image to grayscale image
for i in range(row):
    for j in range(col):
        # Find the average of the BGR pixel values
        img[i, j] = min((sum(img[i, j]) //3)+10,255)   #to reduce contrast as i was already getting image with levels widely spread
plt.imshow(img)
```

histogram

```python
ni=[0]*256
for i in range(row):
    for j in range(col):
        # Find the average of the BGR pixel values
        ni[sum(img[i, j])//3]+=1
hi=[i/(row*col) for i in ni]
plt.plot(hi)
plt.show()
```

cumulative histogram

```python
ci=[0]*255
for i in range(0,255):
  for j in range(0,i+1):
    ci[i]+=hi[j]

plt.plot(ci)
plt.show()

```

brighten the image

```python
nb=np.copy(img)
a =1.8

for i in range(row): #Spans the x-range
    for j in range(col): #Spans the y-range
        nb[i,j]=[ min(a*px,255) for px in nb[i,j]]
plt.imshow(nb)
```

contrast adjustment

```python
r_max=0
r_min=255
for i in range(row):
    for j in range(col):
        pix = sum(img[i, j]) // 3  # Calculate the pixel value as the average of BGR channels
        r_max = max(r_max, pix)
for i in range(row):
    for j in range(col):
        pix = sum(img[i, j]) // 3  # Calculate the pixel value as the average of BGR channels
        r_min = min(r_min, pix)

print("Maximum Pixel Value:", r_max)
print("Min Pixel Value:", r_min)
```

```python
for i in range(row):
    for j in range(col):
      r=sum(img[i, j]) // 3
      img[i, j] = 255*(r-r_min)//(r_max-r_min)  #automatic contrast adjustment
plt.imshow(img)
```
